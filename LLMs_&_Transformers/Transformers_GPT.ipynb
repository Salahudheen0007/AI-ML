```python
from transformers import pipeline
```

## Text Classification

```python
#load  a pretrained sentiment analysis pipeline
classifier = pipeline("sentiment-analysis")
```

```python
result = classifier("I didnt  enjoyed the batman movie")
print(result)
```

## Text Summerization

```python
summarizer = pipeline("summarization",model = "t5-small")
text = """

Sample paragraph for text summarization:
Artificial intelligence (AI) and automation are changing industries and the job market, bringing potential benefits such as increased efficiency and innovation, alongside challenges like job displacement and ethical issues. A balanced strategy is required to maximize AI's advantages while minimizing risks.
"""
summary = summarizer(text,max_length = 40,min_length =10,do_sample = False)
print(summary[0]['summary_text'])
```

## Text Translation

```python
from transformers import MarianTokenizer,MarianMTModel
```

```python
model_name = 'Helsinki-NLP/opus-mt-en-ml'
```

```python
#load tokenizer and model
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
#input
english_text = ["My name is shebin"]

# tokenize
inputs = tokenizer(english_text,return_tensors="pt",padding=True)

#Generate translation
translated = model.generate(**inputs)

#decode translation
malayalam_translation = tokenizer.decode(translated[0],skip_special_tokens = True)

print("English",english_text[0])
print("malayalam",malayalam_translation)
```

## Question Answering(BERT/DistilBERT)

```python
qa_pipeline = pipeline("question-answering")
```

```python
context = """
Artificial intelligence (AI) and automation are changing industries and the job market, bringing potential benefits such as increased efficiency and innovation, alongside challenges like job displacement and ethical issues. A balanced strategy is required to maximize AI's advantages while minimizing risks.
"""
result = qa_pipeline(question = "AI and automation are changing what?",context = context)
print(result)
```

## Text Generation (GPT-2)

```python
from transformers import set_seed
```

```python
generator = pipeline("text-generation",model = 'gpt2')
set_seed(42)
```

```python
output = generator("In the future ,I will",max_length = 10,num_return_sequences = 1)

print(output[0]['generated_text'])
```
