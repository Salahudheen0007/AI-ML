```python
import pandas as pd
movie_review = pd.read_csv("movie_review.csv")
```

```python
movie_review
```

```python
movies_titles = movie_review['Title']
ratings = movie_review.drop(columns=['Title'])
```

```python
movies_titles
```

```python
ratings
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_similarity = cosine_similarity(ratings)
movie_sim_df = pd.DataFrame(movie_similarity,index = movies_titles,columns=movies_titles)
movie_sim_df
```

```python
user_similarity = cosine_similarity(ratings.T)
user_sim_df = pd.DataFrame(user_similarity,index = ratings.columns,columns=ratings.columns)
user_sim_df
```

```python
def similar_users(user,top_n = 3):
  if user not in user_sim_df.columns:
    return f"user '{user}' not found"
  sims = user_sim_df[user].sort_values(ascending=False).drop(user)
  return sims.head(top_n)

def similar_movies(movie,top_n = 3):
  if movie not in movie_sim_df.columns:
    return f"Movie '{movie}' not found!"
  sims = movie_sim_df[movie].sort_values(ascending=False).drop(movie)
  return sims.head(top_n)
```

```python
print("Users similar to Rohit:")
print(similar_users("Rohit", top_n=3))
```

```python
print("\nMovies similar to L2: Empuraan:")
print(similar_movies("L2: Empuraan", top_n=3))
```
