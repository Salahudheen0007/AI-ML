```python
import matplotlib.pyplot as plt

```

# line plot

```python
x = [0,1,2,3,4,5]  # it should be in ascending order
y = [0,3,2,22,22,10]
plt.plot(x,y)

plt.xlabel("days")
plt.ylabel("counts")
plt.xlim(1,5)
plt.ylim(-0.5,0.5)

plt.show()
```

```python
import numpy as np

f = 50
x = np.arange(0,400)
y = np.cos(2*np.pi*f*x/2000)
plt.plot(x,y)
plt.xlim(300,400)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

```python
x = [0,1,2,3,4,5]
y1 = [0,3,2,22,22,10]
y2 = [1,7,5,4,3,8]
plt.plot(x,y1, c = "y" , label = "location 1")
plt.plot(x,y2, c = "r" , label = "location 2")
plt.xlabel("days")
plt.ylabel("counts")
plt.legend()
plt.show()
```

# barplot

```python
# barplot


x = ["Monday","Tuesday","Wednesday","Thursday","Friday"]
item_1_sales = [1,2,3,4,5]
plt.bar(x,item_1_sales, label = "item1")
plt.xlabel("Days")
plt.ylabel("Counts")
plt.savefig("barchart.png",dpi =300)
plt.show()
```

# barplot

```python
# barplot


x = ["Monday","Tuesday","Wednesday","Thursday","Friday"]
item_1_sales = [1,2,3,4,5]
item_2_sales = [3,4,5,6,5]
plt.bar(x,item_1_sales, color = "r" ,label = "item1")
plt.bar(x,item_2_sales, color = "b" ,alpha = 0.5,label = "item2")  # here alpha represents the transparency
plt.xlabel("Days")
plt.ylabel("Counts")
plt.legend(loc = "upper left")
plt.savefig("barchart1.png",dpi =300)
plt.show()
```

```python
countries = ["india","Us","UK","UAE"]
population = [123,20,30,40]
plt.pie(population, labels = countries, autopct = "%1.2f%%")
plt.show()
```

# scatter plot

```python
# scatter plot

import pandas as pd
data = pd.read_csv("auto-mpg.csv")
plt.figure(figsize=(6,3.5))
plt.scatter(data["weight"],data["mpg"],marker = "o")
plt.title("MPG vs WEIGHT")
plt.xlabel("weight")
plt.ylabel("MPG")
plt.show()
```

# sub figures



```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(data["weight"],data["mpg"],c = "k")
plt.ylabel("mpg")
plt.xlabel("weight")
plt.subplot(1,2,2)
plt.hist(data["displacement"])
plt.ylabel("displacement")
plt.xlabel("mpg")
plt.show()
```

```python
import numpy as np
f = 50
x = np.arange(0,400)
y1 = np.cos(2*np.pi*f*x/2000)
y2 = np.sin(2*np.pi*f*x/2000)
plt.plot(x,y1,c = "k",label = "$y = cos(x)$")
plt.plot(x,y2,c = "b",alpha = 0.8,label = "$y = sin(x)$")
plt.legend(loc = "upper left")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.text(100,1,"some waves")
plt.grid()
plt.show()
```

```python
# multiple axis graphs
fig1,ax1 = plt.subplots()
x = np.arange(0,80)
y1 = 100*np.cos(2*np.pi*f*x/2000)
y2 = np.sin(2*np.pi*f*x/2000)
ax2 =  ax1.twinx()
ax1.plot(x,y1,label = "signal 1",c = 'k',linestyle = '--')
ax2.plot(x,y2,label = "signal 2",c = 'b')
ax1.legend()
ax2.legend()
plt.show()
```
