```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x,y,label = '2x',color = 'red',linestyle = '--',linewidth = 5,marker = '.',markersize = 20,markeredgecolor = "b")
plt.title("our 1st graph",fontdict={'fontname' : 'Comic Sans Ms','fontsize' : 20})
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()  
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
#resize the graph
plt.figure(figsize=(2,3),dpi = 300)
plt.plot(x,y, 'b^--',label = '2x',)
plt.title("our 1st graph",fontdict={'fontname' : 'Comic Sans Ms','fontsize' : 20})

x2 = np.arange(0,4.5,.5)
# plt.plot(x2,x2**2,'r',label = 'x^2')
plt.plot(x2[:6],x2[:6]**2,'r',label = 'x^2')
plt.plot(x2[5:],x2[5:]**2,'r--',)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8])
plt.legend()
plt.savefig('1st.png',dpi = 300)
plt.show()  
```

# barplot

```python
labels = ['A','B','C']
values = [1,4,2]
bars = plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.show()
```
