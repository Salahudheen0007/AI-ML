# heat maps

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
heat_data = np.array([[0,0,4,1],[1,0,4,1]])
sns.heatmap(heat_data,annot = True)
```

```python
data = pd.read_csv("auto-mpg.csv")
data_new = data.drop(["car name","horsepower"],axis = 1)
sns.heatmap(data_new.corr(),annot = True)
plt.xticks(rotation = 45)
plt.show()
```

```python
data = pd.read_csv("auto-mpg.csv")
data_new = data.drop(["car name","horsepower"],axis = 1)
corr_mat = data_new.corr()
mask = np.triu(np.ones_like(corr_mat,dtype = bool))
sns.heatmap(corr_mat,mask = mask,annot = True,cmap = "viridis")
plt.xticks(rotation = 45)
plt.show()
```

```python
data = pd.read_csv("auto-mpg.csv")
plt.figure(figsize=(16,15))
plt.subplot(2,2,1)

year_counts = data['model year'].value_counts().sort_index()
plt.bar(year_counts.index, year_counts.values, color='orange')
plt.xlabel('Model Year')
plt.ylabel('Number of Cars')
plt.title('Number of Cars by Model Year')
plt.xticks(year_counts.index)

# plt.tight_layout()
plt.subplot(2,2,2)

plt.scatter(data["weight"],data["acceleration"])
plt.ylabel("acceleration")
plt.xlabel("weight")
plt.title("WEIGHT VS ACCELERATION")
plt.subplot(2,2,3)

num_df = data.select_dtypes(include = np.number)
plt.boxplot(num_df)
plt.title("BOXPLOT")
plt.subplot(2,2,4)
data_new = data.drop(["car name","horsepower"],axis = 1)
sns.heatmap(data_new.corr(),annot = True)
plt.xticks(rotation = 45)
plt.title("HEATMAP OF COORELATIONs")

plt.savefig("dashboard_salahudheen.png")
```
