# lineplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
plt.Figure(figsize=(8,5))

plt.title("Gas prices over time in USD",fontdict={'fontweight' : 'bold','fontsize' : 18})
plt.plot(gas.Year,gas.USA, 'b.-',label = "USA")
plt.plot(gas.Year,gas.Canada,'r.-', label = "Canada")
plt.plot(gas['Year'],gas['South Korea'],'g.-',label = "SC")
# for country in gas:
#     if country != "Year":
#         plt.plot(gas.Year,gas[country],marker = '.')
print(gas.Year[::3])
plt.xticks(gas.Year[::3].tolist()+[2011])


plt.legend()
plt.savefig("Gas pricses .png")
plt.show()
```

# load fifa data


```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head()
```

# Histogram

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins = bins,color = 'b')
plt.ylabel("no of playesr")
plt.xlabel("skill level")
plt.title("distribution of players skill")
plt.show()
```

```python
fifa.head()
labels = ["left","right"]
color = ["#ababab","#abacde"]
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left,right],colors = color,labels = labels,autopct='%.2f %%')
```

```python
fifa.Weight
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight] 
plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight  < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >=175 )  &  (fifa.Weight <200)].count()[0]
heavy = fifa[fifa.Weight > 200].count()[0]

Weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['under 125','125-150','150-175','175-200','above 200']
explode = (.4,.2,0,0,.4)
plt.pie(Weights,labels = labels,explode = explode)
plt.show()
```

# Box plot

```python
fifa.head()
```

```python
plt.style.use('default')
barcelona = fifa[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa[fifa.Club == "Real Madrid"]["Overall"]
revs = fifa[fifa.Club == "New England Revolution"]['Overall']
labels = ['FC Barcelona','Real Madrid','New England Revolution']
boxes = plt.boxplot([barcelona, madrid,revs],labels = labels,patch_artist=True)
for box in boxes['boxes']:
    box.set(color = 'b') # edge color
    box.set(facecolor = 'y')
plt.title("professional football team comparion")
plt.boxplot([barcelona, madrid,revs],labels = labels)
plt.show()

```
