## Fourier decomposition/Transform

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd

```

```python
audio_data,sr = librosa.load("abc.mp3",sr = None,duration = 2)
ipd.Audio(audio_data,rate = sr)
```

```python
audio_data.shape
```

```python
plt.figure(figsize=(6,4))
librosa.display.waveshow(audio_data,sr=sr)

plt.title("Time domain Wavefome")
plt.xlabel("time (s)")
plt.ylabel("Amplitude")
plt.show()
```

## FFT spectrum

```python
fft_spectrum = np.fft.fft(audio_data)
frequencies = np.fft.fftfreq(len(audio_data),1/sr)
plt.figure(figsize=(6,4))
plt.plot(frequencies[:len(frequencies)//2],np.abs(fft_spectrum[:len(frequencies)//2]))
plt.title("frequenct domain spectrum")
plt.xlabel("Freq (Hz)")
plt.ylabel("magnitusde")
plt.show()
```

```python
# Load a 2-second excerpt from a built-in audio file
y, sr = librosa.load(librosa.ex('libri2'), sr=None, duration=2)

# To confirm duration
print(f"Sample rate: {sr}, Duration: {len(y)/sr:.2f} seconds")
ipd.Audio(y, rate=sr)
```

## stft magnitude


```python
stft_spectrogram = librosa.stft(audio_data)
stft_magnitude = np.abs(stft_spectrogram)
stft_db = librosa.amplitude_to_db(stft_magnitude,ref = np.max)
plt.figure(figsize=(6,4))

librosa.display.specshow(stft_db,sr = sr,x_axis = 'time',y_axis="hz")

plt.colorbar(format = '%+2.0f dB')
plt.title("STFT spectrogram")
plt.xlabel("time")
plt.ylabel("Freq")
plt.show()
```

## MFCC (for speech)

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

y, sr = librosa.load("abc.mp3")
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```
