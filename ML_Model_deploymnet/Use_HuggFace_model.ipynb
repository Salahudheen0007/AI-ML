```python
from huggingface_hub import hf_hub_download
import joblib
```

```python
USERNAME = "Salahudheen007"
REPO_NAME = "iris-rf-model"

repo_id = f"{USERNAME}/{REPO_NAME}"
```

```python
print(f"Downloading your model from: {repo_id}")
# This is the line that downloads your file
path = hf_hub_download(repo_id=repo_id, filename="rf_iris.joblib")
```

```python
model_from_hub = joblib.load(path)
print("Model loaded successfully!")
```

```python
sample_data = [5.1, 3.5, 1.4, 0.2]

prediction = model_from_hub.predict([sample_data])
species_names = ['Setosa', 'Versicolor', 'Virginica']
prediction_index = prediction[0]

print(f"The model predicts species: {species_names[prediction[0]]}")
```
