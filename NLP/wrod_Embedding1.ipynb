```python
  sentences = [
    ["the", "glass", "of", "milk"],
    ["the", "glass", "of", "juice"],
    ["the", "cup", "of", "tea"],
    ["i", "am", "a", "good", "boy"],
    ["i", "am", "a", "good", "developer"],
    ["understand", "the", "meaning", "of", "words"],
    ["your", "videos", "are", "good"]
]
```

```python
# training
model = Word2Vec(sentences,vector_size=10,window=5,min_count=1,workers=4)
```

```python
print("vector for good",model.wv['good'])
```

```python
print("most similar to word good",model.wv.most_similar('good'))
```

```python
words = model.wv.key_to_index
print(list(words.keys())[:10])
```
