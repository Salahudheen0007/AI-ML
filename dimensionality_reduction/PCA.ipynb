```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Iris.csv')
print(data.shape)
```

```python
data.head()

```

```python
y = data['Species']
X = data.drop(['Species','Id'],axis = 1)
X

from sklearn.preprocessing import StandardScaler

sc  = StandardScaler()
scaled_data  = sc.fit_transform(X)
scaled_data
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit_transform(scaled_data)
explained_variance = pca.explained_variance_ratio_

explained_variance
```

```python
plt.bar(range(1,len(explained_variance)+1),explained_variance)
plt.xlabel("principle components")
plt.ylabel("Explained_variance")
```

```python
pca1 = PCA(n_components=2)
pca1.fit_transform(scaled_data)

X_new = pca1.fit_transform(scaled_data)

explained_variance1 = pca1.explained_variance_ratio_
print(explained_variance1)
print(X_new)

# pca1.get_covariance()
```

```python
plt.bar(range(1,3),explained_variance1)
plt.xlabel("principle components")
plt.ylabel("explained var")
```

```python

plt.scatter(X_new[:,0],X_new[:,1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Projection")
plt.show()
```

```python
print(f"percentage sum of explained variance: {np.round(np.sum(explained_variance1)*100,2)}")
```
